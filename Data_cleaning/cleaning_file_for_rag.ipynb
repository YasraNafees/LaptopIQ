```python
def clean_weight(weight):
    if isinstance(weight, str):
        weight = weight.replace('kg', '').strip()
    try:
        return float(weight)
    except:
        return None
```

```python
df.head()
```

```python
def clean_os(os_name):
    if pd.isna(os_name): return "Unknown"
    os_name = os_name.strip()
    # Standardizing names for better RAG retrieval
    if 'Mac' in os_name or 'macOS' in os_name:
        return 'macOS'
    elif 'Windows' in os_name:
        return 'Windows' # 'Windows 10' ko 'Windows' bhi saktay ho agar generic chahiye
    elif 'Linux' in os_name or 'Ubuntu' in os_name:
        return 'Linux'
    elif 'Android' in os_name:
        return 'Android'
    elif 'No OS' in os_name:
        return 'No OS'
    else:
        return os_name
```

```python
df.head()
```

```python
def clean_memory(memory):
    if pd.isna(memory): return ""
    return memory.strip().replace("  ", " ")
```

```python
def create_document_text(row):
    """
    RAG ke liye ek sentence generate karna taake search easy ho.
    """
    return (f"Company: {row['Company']}, "
            f"Product: {row['Product']}, "
            f"CPU:  {row['CPU_Type']} , "
            f"RAM: {row['RAM (GB)']} GB, "
            f"Memory: {row['Memory']}, "
            f"OS: {row['OpSys']}, "
            f"Weight: {row['Weight (kg)']} kg, "
            f"Price: {row['Price (USD)']} USD, ")
```

```python
df.columns = df.columns.str.strip()

df['Weight (kg)'] = df['Weight (kg)'].apply(clean_weight)
df['OpSys'] = df['OpSys'].apply(clean_os)
df['Memory'] = df['Memory'].apply(clean_memory)
```

```python
df['Product'] = df['Product'].str.split('(').str[0].str.strip()
```

```python
df = df.sample(n=200).reset_index(drop=True)
```
